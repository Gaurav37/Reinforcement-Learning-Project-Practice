```python
# -*- coding: utf-8 -*-
"""
Created on Fri Oct  4 16:16:45 2019

@author: gaura
"""

import numpy as np
import matplotlib.pyplot as plt
#rom gym.envs.toy_text.frozen_lake import FrozenLakeEnv
def gamblers_possible_actions(s):
    """
    Args:
        s: Current state, i.e. amount of money available
    Returns:
        An iterable of actions which can be taken from that state
    """
    return range(1, min(s, 100 - s) + 1)
    
    
def gamblers_value_iteration(p_h, theta=1e-4, gamma=1.0):
    """
    Args:
        p_h: Probability of coin flip being heads
        theta: Threshold used to determine accuracy of estimation
        gamma: Discount factor
    Returns:
        A tuple (value function, policy)
    """

    V = np.zeros(101)
    policy = np.zeros(100, dtype=np.int)
    
    var = 1
    while var == 1 :
        delta = 0
        for i in range(len(V)):
            print("i=", i)
            Vlocal=V[i]
            print("Vlocal", Vlocal)
            reward = 0
            action_options = gamblers_possible_actions(i)
            print("actions available", action_options)
            if i==100:
                reward=1
            Estimated_values = np.zeros(len(action_options))
            print("Estimated values", Estimated_values)
#           if Estimated_value.size>0:
            for action in action_options:
                Estimated_values[action-1]=p_h*(reward + V[ i + action])+(1-p_h)*( V[i-action] ) #no reward for 1-p_h 
                
            if len(Estimated_values)>0:
                optimal=np.argmax(Estimated_values)
                print("", optimal)
                V[i]=Estimated_values[optimal]
                policy[i]=optimal#+1
                print("optimal policy action")
#             policy[i]=V[i]
            else:
                V[i]=reward
                print("no estimates")
            
            delta=max(delta, abs(Vlocal-V[i]) )
        if delta<theta: 
            break
    return V,policy

    V,policy=gamblers_value_iteration(0.4)
    print ("V", V, policy)
```

```python
V, policy=gamblers_value_iteration(0.4)
```
